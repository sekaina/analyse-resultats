```python
def my_to_datetime(date_str):
    if date_str[8:10] != '24':
        return pd.to_datetime(date_str, format=' %m/%d  %H:%M:%S')
    date_str = date_str[0:8] + '00' + date_str[10:]
    return pd.to_datetime(date_str, format=' %m/%d  %H:%M:%S') + dt.timedelta(days=1)

print(my_to_datetime(' 04/10  24:00:00'))
```

```python
df_nomass=pd.read_csv("./modelNoMASS/IDM_NoMASS.csv",index_col="Date/Time", parse_dates=True, date_parser=my_to_datetime)
df_deter=pd.read_csv("C:/users/elkhatts/desktop/IDM/IDM.csv",index_col="Date/Time", parse_dates=True, date_parser=my_to_datetime)
```
